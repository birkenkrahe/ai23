```r
library(ggplot2)
ggplot(mtcars, aes(x=wt, y=mpg)) +
  geom_point(aes(color=cyl), size=3) +
  labs(title='Scatterplot of Miles-per-Gallon vs. Weight',
       x='Weight (in 1000 lbs)',
       y='Miles-per-Gallon',
       color='Number of Cylinders') +
  theme_minimal()
```

```r
plot(mtcars$wt, mtcars$mpg,
     main='Scatterplot of Miles-per-Gallon vs. Weight',
     xlab='Weight (in 1000 lbs)',
     ylab='Miles-per-Gallon',
     col=mtcars$cyl,
     pch=19, cex=1.5)
legend('topright', legend=unique(mtcars$cyl), fill=unique(mtcars$cyl), title='Number of Cylinders')
```

```r
linear_model <- lm(mpg ~ wt, data=mtcars)
plot(mtcars$wt, mtcars$mpg,
     main='Scatterplot of Miles-per-Gallon vs. Weight with Trendline',
     xlab='Weight (in 1000 lbs)',
     ylab='Miles-per-Gallon',
     col=mtcars$cyl,
     pch=19, cex=1.5)
abline(linear_model, col='red', lwd=2)
legend('topright', legend=unique(mtcars$cyl), fill=unique(mtcars$cyl), title='Number of Cylinders')
```

```r
library(ggplot2)
ggplot(mtcars, aes(x=wt, y=mpg, color=cyl, size=hp)) +
  geom_point(alpha=0.7) +
  scale_color_brewer(palette='Dark2') +
  labs(title='Miles-per-Gallon vs. Weight',
       subtitle='Colored by Number of Cylinders, Sized by Horsepower',
       x='Weight (in 1000 lbs)',
       y='Miles-per-Gallon',
       color='Number of Cylinders',
       size='Horsepower') +
  theme_minimal() +
  theme(legend.position='bottom')
```

```r
ggplot(mtcars, aes(x=wt, y=mpg, color=as.factor(cyl), size=hp)) +
  geom_point(alpha=0.7) +
  scale_color_brewer(palette='Dark2') +
  labs(title='Miles-per-Gallon vs. Weight',
       subtitle='Colored by Number of Cylinders, Sized by Horsepower',
       x='Weight (in 1000 lbs)',
       y='Miles-per-Gallon',
       color='Number of Cylinders',
       size='Horsepower') +
  theme_minimal() +
  theme(legend.position='bottom')
```
